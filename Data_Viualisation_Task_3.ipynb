```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/householdtask3.csv")
```

```python
df.head(10)
```

```python
plt.scatter(df['expenditure'],df['income'])
plt.title("Scatter Plot")
plt.xlabel('expenditure')
plt.ylabel('income')
```

```python
plt.scatter(df['expenditure'],df['income'])
plt.title("Scatter Plot")
plt.xlabel('expenditure')
plt.ylabel('income')
plt.colorbar()
plt.show()
```

```python
plt.plot(df['expenditure'])
plt.plot(df['income'])

plt.title("Scatter Plot")
plt.xlabel('expenditure')
plt.ylabel('income')
plt.show()
```

```python
plt.bar(df['prop_hhs'],df['tot_hhs'])
plt.title("Bar Chart")
plt.xlabel('prop_hhs')
plt.ylabel('tot_hhs')
df[['prop_hhs', 'tot_hhs']].hist(bins=20, figsize=(15, 10))
plt.show()
```
